#Removing Backgrounds from Images


`rembg` is an open-source Python library used for removing backgrounds from images. It simplifies the process of extracting the main subject from an image and removing or making the background transparent. The library is built on deep learning models, specifically optimized for high accuracy in distinguishing the foreground (subject) from the background, making it a useful tool for various applications such as image editing, product photography, and visual content creation.

##Background Removal

Automatically detects and removes backgrounds from images, leaving just the subject with a transparent background. Handles complex backgrounds effectively, even in images with diverse scenes.

##CLI

`rembg` can be used directly from the command line to process images without writing code.

`rembg i input.jpg output.png`

##Python Library
`rembg` can be integrated into `Python` projects, making it easy to incorporate background removal into automated workflows.

### Download libraries


```python
from rembg import remove
from PIL import Image
import matplotlib.pyplot as plt
```

### Import data

```python
!wget https://raw.githubusercontent.com/tahamsi/computer-vision/refs/heads/main/images/duck.jpg
```

```python
input_image = Image.open('duck.jpg')
plt.imshow(input_image)
```

### Remove the background

```python
output_image = remove(input_image)
output_image.save('output.png')
plt.imshow(output_image)
```

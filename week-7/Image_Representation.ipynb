## Image Representation
Images are essentially grids of pixels, which can be represented as 2D arrays, making image manipulation more accessible. Each pixel's coordinates indicate its position within the image. To represent color, various schemes exist, but a common choice is the RGB scheme. In this scheme, an additional dimension is added to the grid to represent the color channels (Red, Green, and Blue), where each channel holds intensity values for its respective color.

## Reading an Image
In Python, you can read images using libraries such as Pillow (PIL), OpenCV, or Matplotlib. Each library offers unique tools for image handling, from loading and manipulating images to displaying and saving them.

**Using Pillow (PIL)**

The Pillow library is simple and effective for basic image processing.

```python
from PIL import Image
from IPython.display import display

# Load an image
image = Image.open("peel.jpg")

# Get image details
print(image.size)       # (width, height)
print(image.mode)       # Color mode, e.g., RGB, L, etc.
```

**Using Matplotlib**

Matplotlib’s pyplot module is useful for displaying images in a Jupyter Notebook or other interactive environments.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generating a random array with integer values between 0 and 255.
By default, plt.imshow() displays images as a heatmap.

```python
random_array = np.random.randint(0, 255, (200,200))
print(random_array.shape)

plt.imshow(random_array)
```

Passing cmap='gray' displays the image in grayscale.

```python
plt.imshow(random_array, cmap='gray')
```

*plt.imread()* reads image files and converts them into an ndarray.

```python
image = plt.imread('peel_gs.jpg')

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot on the first subplot
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original image')

# Plot on the second subplot
axs[1].imshow(image)
axs[1].set_title('Heatmap')
```

Grayscal images are 2-D matrixes or 2-D arrays

```python
print(image.shape)
```

RGB images are stored in 3-D tensors, with the last dimension representing the color channels.

```python
image = plt.imread('peel.jpg')
print(image.shape)
```

By slicing and indexing (specifying the address for each channel), you can access them individually.

```python
fig, axs = plt.subplots(1, 4, figsize=(12, 5))

axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original image')

axs[1].imshow(image[:,:,0])
axs[1].set_title('Red channel')

axs[2].imshow(image[:,:,1])
axs[2].set_title('Green channel')

axs[3].imshow(image[:,:,2])
axs[3].set_title('Blue channel')
```
